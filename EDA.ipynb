```python
# Verify null to unique rate
nulls = df.isnull().sum().tolist()
uniques = df.nunique().tolist()

division = [unique / null if null != 0 else 0 for unique, null in zip(uniques, nulls)]
pd.DataFrame(division, index = df.columns, columns = ['unique_to_null_rate'])
```
